# EDA

Let's find first the period of time used in this study

Now let's find how many times a team has reach the 1st position of the ranking between 1993 and 2018

```python
# gets the amount of times a confederation has reach the 1st position
leadershipsPerTeam = dataSet[(dataSet['rank'] == 1)]['country_full'].value_counts()
leadershipsPerTeam
```

Visualizing the leaderships per team as a bar chart:

```python
plot.title('Leaderships per team')
x = leadershipsPerTeam.keys().tolist()
y = leadershipsPerTeam.tolist()
plot.bar(x, y)
```

From the confederation point of view:

```python
# gets the amount of times a confederation has reach the 1st position
polesPerConfederation = dataSet[(dataSet['rank'] == 1)]['confederation'].value_counts()
polesPerConfederation
```

Even only 2 COMNEBOL (South America) teams have reach the 1st position, against 6 European (UEFA) teams, COMNEBOL is leading the total count  by far.

Now let's find about the best positions reached by Central and North American teams (CONCACAF)

```python
# get all concacaf teams
concacafTeams = dataSet[(dataSet['confederation'] == 'CONCACAF')]
concacafTeams = concacafTeams.sort_values(by='rank', ascending=True)

bestPositionByTeam = pandas.DataFrame(columns=['country', 'rank'])

# get best position reach by each country
for team in concacafTeams['country_full'].unique():
    pos = concacafTeams[(concacafTeams['country_full'] == team)]['rank'].min()
    bestPositionByTeam.loc[len(bestPositionByTeam)] = [team, pos]
    
bestPositionByTeam.head(15)
```

From the earned total_points point of view:

```python
pointsByTeam = pandas.DataFrame(columns=['country', 'points'])

concacafTeams = dataSet[(dataSet['confederation'] == 'CONCACAF')]
concacafTeams = concacafTeams.sort_values(by='rank', ascending=True)

# get total points by team
for team in concacafTeams['country_full'].unique():
    points = concacafTeams[(concacafTeams['country_full'] == team)]['total_points'].sum()
    pointsByTeam.loc[len(pointsByTeam)] = [team, points]
    
pointsByTeam.sort_values(by='points', ascending=False).head(5)
```

```python
# show data as bar chart
plot.title('Total point earned from 1993 to 2018 per team')
x = pointsByTeam['country'][:5]
y = pointsByTeam['points'][:5]
plot.bar(x, y)
```
